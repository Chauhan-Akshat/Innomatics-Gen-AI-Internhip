Task 1: Conceptual Understanding 
1. What is the difference between "Love" and "love" in NLP? 
In NLP, text is case-sensitive by default, which means "Love" and "love" are treated as completely different words. This can mess things up because it increases the number of unique words in the dataset and makes the model less efficient. To avoid this, we usually convert everything to lowercase during preprocessing so both "Love" and "love" are treated the same. 
2. What happens if stopwords are not removed? 
If we don’t remove stopwords, the dataset ends up full of common words like "is", "the", and "and" that don’t really add much meaning. This makes the data more bulky and noisy. Due to which, models take more time to process the data and may not perform as well since they’re focusing on useless words instead of meaningful ones. 
3. Provide two real-world scenarios where removing stopwords can be harmful. 
In sentiment analysis, removing words like "not" can completely change the meaning; for example, "not good" would become just "good". This is obviously wrong! In question-answer systems, removing words such as who, what or where takes away the main intent of the question making it harder for a model to understand what’s being asked. 
4. Explain the difference between stemming and lemmatization.  
Stemming: A rule-based approach where we remove prefixes or suffixes to get to some base form of a word without considering its context . Examples: "running" → run; 
"studies" → studi 
Lemmatization: An advanced process that takes into account the context of a word before converting it into its actual dictionary base form so that output is usually a proper word.

```python
import re

def preprocess_text(text):
    try:
        if not isinstance(text, str) or text.strip() == "":
            return [], ""

        text = text.lower()

        text = re.sub(r"http\S+|www\S+|https\S+|\S+@\S+", "", text)

        text = re.sub(r"\d+", "", text)

        text = re.sub(r'(.)\1{2,}', r'\1', text)

        text = re.sub(r'[^a-zA-Z\s]', '', text)

        text = re.sub(r'\s+', ' ', text).strip()

        tokens = text.split()

        tokens = [word for word in tokens if len(word) > 2 or word in ['no', 'not']]

        clean_sentence = " ".join(tokens)

        return tokens, clean_sentence

    except Exception as e:
        return [], ""
```

```python
test_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

for sentence in test_sentences:
    tokens, clean = preprocess_text(sentence)
    print("\nOriginal:", sentence)
    print("Tokens:", tokens)
    print("Clean Sentence:", clean)
```

```python
def token_analysis(tokens):
    if len(tokens) == 0:
        return 0, 0, 0

    total = len(tokens)
    unique = len(set(tokens))
    avg_len = sum(len(word) for word in tokens) / total

    return total, unique, round(avg_len, 2)
for sentence in test_sentences:
    tokens, _ = preprocess_text(sentence)
    total, unique, avg = token_analysis(tokens)
    
    print("\nSentence:", sentence)
    print("Total:", total, "Unique:", unique, "Avg Length:", avg)
```

```python
from collections import Counter

all_tokens = []

for sentence in test_sentences:
    tokens, _ = preprocess_text(sentence)
    all_tokens.extend(tokens)

counter = Counter(all_tokens)

print("Top 10 words:", counter.most_common(10))
print("Least 5 words:", counter.most_common()[-5:])
```

```python
def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens, clean = preprocess_text(text)
        all_tokens.extend(tokens)
        clean_sentences.append(clean)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }
```

```python
edge_cases = ["", "😂😂😂😂", "123456"]

for case in edge_cases:
    tokens, clean = preprocess_text(case)
    print("\nInput:", case)
    print("Tokens:", tokens)
    print("Clean:", clean)
```
