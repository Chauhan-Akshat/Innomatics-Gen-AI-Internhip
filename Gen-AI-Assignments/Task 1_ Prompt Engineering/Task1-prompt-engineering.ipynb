```python
# Import ChatPromptTemplate for creating role-based chat prompts
from langchain_core.prompts import ChatPromptTemplate


# Create a chat prompt template with system and user roles
# System defines the role (teacher/interviewer/motivator)
# User provides the actual question/topic
chat_template = ChatPromptTemplate.from_messages([
    ("system", "you are a {role}"),
    ("user", "Explain {topic}")
])


# Define different roles to generate varied responses
roles = ["teacher", "interviewer", "motivator"]

# Topic to be explained
topic = "Neural Networks"


# Combine chat template with model using LCEL
chain = chat_template | model


# Loop through each role and generate role-specific responses
for role in roles:

    # Invoke the chain with current role and topic
    response = chain.invoke({"role": role, "topic": topic})

    # Print the generated response
    print(response.content)

    # Separator for readability
    print("-" * 60)
```

```python
# Function to validate audience and tone inputs
# Assigns default values if invalid inputs are provided
def validate_inputs(audience, tone):

    # Define allowed values for audience and tone
    valid_audience = ["beginner", "intermediate", "expert"]
    valid_tone = ["formal", "casual", "fun"]

    # Convert inputs to lowercase for consistency
    audience = audience.lower()
    tone = tone.lower()

    # Validate audience
    # If not valid, assign default value
    if audience not in valid_audience:
        print("Invalid audience! Defaulting to beginner.")
        audience = "beginner"

    # Validate tone
    # If not valid, assign default value
    if tone not in valid_tone:
        print("Invalid tone! Defaulting to formal.")
        tone = "formal"

    # Return validated (or corrected) values
    return audience, tone


# Testing the function with valid input
print(validate_inputs("beginner", "fun"))

# Testing the function with invalid input
print(validate_inputs("kids", "cool"))
```

**Task 6: Prompt Generator App**
* Create function:
generate_prompt(topic, audience, tone, style)

* Requirements:
  - Use PromptTemplate
  - Styles: teaching, interview, storytelling
* Example Output:
Explain Neural Networks for beginners in a fun storytelling style


```python
# Function to dynamically generate prompts based on style
def generate_prompt(topic, audience, tone, style):

    # Step 1: Validate inputs (ensure audience and tone are valid)
    audience, tone = validate_inputs(audience, tone)

    # Step 2: Select appropriate template based on style
    if style == "teaching":
        # Template for explanation in teaching style
        template = PromptTemplate(
            input_variables=["topic", "audience", "tone"],
            template="Explain {topic} for {audience} in a {tone} teaching style"
        )

    elif style == "interview":
        # Template for generating interview questions
        template = PromptTemplate(
            input_variables=["topic", "audience", "tone"],
            template="Generate 3 interview questions about {topic} in a {tone} tone"
        )

    elif style == "storytelling":
        # Template for storytelling-style explanation
        template = PromptTemplate(
            input_variables=["topic", "audience", "tone"],
            template="Explain {topic} as a story for {audience} in a {tone} tone"
        )

    else:
        # Handle invalid style input
        return "Invalid style!"

    # Step 3: Format the selected template with inputs
    prompt = template.format(topic=topic, audience=audience, tone=tone)

    # Return final generated prompt
    return prompt
```

```python
# testing the generate_prompt function

print(generate_prompt("Neural Networks", "beginner", "fun", "teaching"))
print(generate_prompt("AI", "kids", "cool", "storytelling"))   # invalid → will auto-fix
print(generate_prompt("Python", "expert", "formal", "interview"))
```

```python
# Import PromptTemplate for creating reusable prompt structure
from langchain_core.prompts import PromptTemplate


# Create a reusable prompt template for dynamic inputs
# Same template will be used with different input values
reusable_template = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template="Explain {topic} for {audience} in a {tone} tone"
)


# Define multiple test cases to demonstrate template reusability
test_cases = [
    {"topic": "AI", "audience": "beginner", "tone": "fun"},
    {"topic": "Python", "audience": "intermediate", "tone": "formal"},
    {"topic": "Data Science", "audience": "expert", "tone": "casual"},
    {"topic": "Machine Learning", "audience": "beginner", "tone": "formal"},
    {"topic": "Deep Learning", "audience": "expert", "tone": "fun"}
]


# Combine template with model to generate real AI responses
chain = reusable_template | model


# Iterate through all test cases
for case in test_cases:

    # Generate formatted prompt (without LLM)
    print(reusable_template.format(**case))

    # Invoke LLM to get actual response
    response = chain.invoke(case)
    print(response.content)

    # Separator for readability
    print("-" * 50)
```
