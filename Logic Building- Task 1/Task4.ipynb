```python
#Mobile Recharge Validation System

AVAILABLE_PLANS = [199, 299, 399, 599]

# This function checks if the recharge is valid or not
def check_recharge(user_amount):
    
    # checking if amount is less than minimum recharge
    if user_amount < 50:
        print("Recharge should be at least ₹50.")
        return False
    
    # checking if amount is in available plans list
    if user_amount not in AVAILABLE_PLANS:
        print("This plan is not available. Choose from:", AVAILABLE_PLANS)
        return False
    return True


# this loop will keep asking until correct recharge is entered
while True:
    try:
        recharge_value = int(input("Enter recharge amount: ₹"))
        
        # calling function to check recharge
        if check_recharge(recharge_value):
            print("Recharge successful ✅")
            break
        else:
            print("Please try again.\n")
    
    except ValueError:
        print("Please enter numbers only.\n")
```

```python
#Inventory Reorder Alert System

inventory = {
    "Laptop": 20,
    "Mouse": 12,
    "Keyboard": 18,
    "Monitor": 10
}

# this function checks stock and tells if reorder is needed
def check_inventory(stock_data):

    for product, quantity in stock_data.items():

        if quantity < 15:
            print(f"{product}: Reorder Alert (Stock = {quantity})")

        else:
            print(f"{product}: Stock OK (Stock = {quantity})")


# calling the function
check_inventory(inventory)
```

```python
#Student Result Processing System

# this function calculates average and tells pass or fail
def process_result(marks):
    
    total = 0

    for m in marks:
        total += m

    average = total / len(marks)

    if average >= 50:
        return "Pass", average
    else:
        return "Fail", average


# example student marks
student_marks = [60, 45, 55, 70]

result, avg = process_result(student_marks)

print("Average Marks:", avg)
print("Result:", result)
```

```python
#Cab Fare Estimator with Retry Option

# this function calculates the cab fare
def fare_calc(km, peak):
    fare = 50 + (12 * km)
    if peak == "yes":
        fare = fare * 1.25
    return fare


# this loop keeps running until correct input is given
while True:
    try:
     
        km = float(input("Distance in km: "))
        peak = input("Peak hour? (yes/no): ").lower()
        total = fare_calc(km, peak)
        print("Total Fare: ₹", total)
        break

    except:
        print("Invalid input, try again\n")
```

```python
#Employee Attendance Eligibility Checker

def check_attendance(att_list):
    
    present = 0 
    for status in att_list:
        
        # if present, increase count
        if status == "P":
            present += 1

    percentage = (present / len(att_list)) * 100

    if percentage >= 75:
        return "Eligible", percentage
    else:
        return "Not Eligible", percentage

attendance = ["P", "A", "P", "P", "A", "P", "P", "P"]

result, percent = check_attendance(attendance)

print("Attendance %:", percent)
print("Status:", result)
```

```python
#Password Strength Checker

def check_password(pwd):
    
    # password should be at least 8 characters
    if len(pwd) < 8:
        return "Weak Password"
    
    has_digit = False      
    has_special = False   

    for ch in pwd:
        # if number found
        if ch.isdigit():
            has_digit = True
        # if special character found
        if ch in "@#$":
            has_special = True
    # password is strong only if both conditions are true
    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password"

password = input("Enter password: ")
print(check_password(password))
```
