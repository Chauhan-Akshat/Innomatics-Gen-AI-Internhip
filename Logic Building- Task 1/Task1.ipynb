```python
#User login check

username="admin"
password="1234"

i_username = input("Enter username: ")
i_password = input("Enter password: ")

if i_username == username and i_password == password:
    print("Login is successful!")
else:
    print("Invalid username or password.")
```

```python
#Pass/ Fail check

marks =  [47, 75, 93, 30, 65]
f_count=0
p_count=0

for mark in marks:
    if mark >= 50:
        p_count=p_count+1
    else:
        f_count=f_count+1
        
print("Number of students passed:", p_count)
print("Number of students failed:", f_count)
```

```python
#Simple Data Cleaner

names = [" Ash ", "ben", " CHARLES "]

cleaned_names = []

for n in names:
    cleaned=n.strip().lower()
    cleaned_names.append(cleaned)
    
print(cleaned_names)
```

```python
#Message Length Analyzer

messages = ["Hey", "Welcome to the platform", "OK"]

for msg in messages:
    length=len(msg)
    print(f"Message: '{msg}' has length: {length}")
    
    if length>10:
        print("FLAGGED: This message is over 10 characters in length")
```

```python
#Error Message Detector

logs = ["INFO", "ERROR", "WARNING", "ERROR"]

e_count=0

for log in logs:
    if log=="ERROR":
        e_count=e_count+1
        
print("Number of errors detected:", e_count) 
```
