```python
#Unique words in a sentence
sentence = input("Enter sentence: ")
words = sentence.split()

unique_words = set(words)

print("Unique words count:", len(unique_words))
print("Unique words:", unique_words)
```

```python
#Highest Salary from employee data
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

highest_employee = max(employees, key=employees.get)
highest_salary = employees[highest_employee]

print("Highest Salary:", highest_employee, "-", highest_salary)
```

```python
#Finding Max and Min values
numbers = [45, 22, 89, 10, 66]

max_num = numbers[0]
min_num = numbers[0]

for num in numbers:
    if num > max_num:
        max_num = num
    if num < min_num:
        min_num = num

print("List:", numbers)
print("Max:", max_num)
print("Min:", min_num)
```

```python
#Counting Products above a price threshold
prices = [450, 1200, 899, 1500, 300]

count = 0

for price in prices:
    if price > 1000:
        count += 1

print("Products above 1000:", count)
```

```python
#Calculating attendance percentage
attendance = ["P", "P", "A", "P", "P"]

present_count = 0

for status in attendance:
    if status == "P":
        present_count += 1

total_days = len(attendance)
percentage = (present_count / total_days) * 100

print("Attendance Percentage:", percentage)
```

```python
#Removing unique phone numbers
phone_numbers = [9876543210, 9123456789, 9876543210]

unique_numbers = set(phone_numbers)

print("Unique phone numbers:", unique_numbers)
```

```python
#Counting character frequency
text = "ppython"

char_count = {}

for char in text:
    if char in char_count:
        char_count[char] += 1
    else:
        char_count[char] = 1

print(char_count)
```

```python
#Checking if a key exists in a dictionary
employees = {
    "Employee": "Raj",
    "Sal": 750000,
    "Dept": "IT"
}

key = "Sal"

if key in employees:
    print("Employee exists")
else:
    print("Employee does not exist")
```

```python
# Calculating average marks
marks = [71, 63, 89, 59]

total = sum(marks)

count = len(marks)

average = total / count

print("Average Marks:", average)
```
