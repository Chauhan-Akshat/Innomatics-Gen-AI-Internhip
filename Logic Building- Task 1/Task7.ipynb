```python
#Smart Light Controller
class SmartLight:
    
    # constructor to set light name and default status
    def __init__(self, name):
        self.name = name
        self.status = "OFF"  
    # function to turn light ON
    def turn_on(self):
        self.status = "ON"
    # function to turn light OFF
    def turn_off(self):
        self.status = "OFF"
    # function to show current light status
    def display_status(self):
        print(f"{self.name} is {self.status}")
# light name input
light_name = "Bedroom Light"
action = "ON"
# creating smart light object
light = SmartLight(light_name)
# checking action and performing operation
if action.upper() == "ON":
    light.turn_on()
else:
    light.turn_off()

light.display_status()
```

```python
#Employee ID Card System
class Employee:
    # constructor to set employee details
    def __init__(self, name, emp_id, department):
        self.name = name
        self.emp_id = emp_id
        self.department = department
    # function to display employee ID card
    def display_id_card(self):
        print("Employee ID Card")
        print(f"Name: {self.name}")
        print(f"ID: {self.emp_id}")
        print(f"Department: {self.department}")

name = "Innomatics Research Labs"
emp_id = "EMP102"
department = "Generative AI"

employee = Employee(name, emp_id, department)

employee.display_id_card()
```

```python
#Mobile Contact Record
class Contact:
    # constructor to set contact details
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone
    # function to display saved contact
    def display_contact(self):
        print("Contact Saved")
        print(f"Name: {self.name}")
        print(f"Phone: {self.phone}")

name = "INNOMATICS RESEARCH LABS"
phone = "9876543210"
# creating contact object
contact = Contact(name, phone)
# displaying contact details
contact.display_contact()
```

```python
#Product Price Tag Generator
class Product:
    # constructor to set product details
    def __init__(self, name, price):
        self.name = name
        self.price = price
    # function to display product price tag
    def display_price_tag(self):
        print(f"Product: {self.name}")
        print(f"Price: ₹{self.price}")

# product details input
product_name = "Headphones"
price = 2499
# creating product object
product = Product(product_name, price)
# displaying price tag
product.display_price_tag()
```

```python
#Movie Rating Display System
class Movie:
    # constructor to set movie details
    def __init__(self, name, rating):
        self.name = name
        self.rating = rating
    # function to display movie information
    def display_movie(self):
        print(f"Movie: {self.name}")
        print(f"Rating: {self.rating} / 5")

movie_name = "Inception"
rating = 4.8
# creating movie object
movie = Movie(movie_name, rating)
# displaying movie details
movie.display_movie()
```

```python
#Delivery Address Manager
class Delivery:
    # constructor to set delivery details
    def __init__(self, customer_name, address):
        self.customer_name = customer_name
        self.address = address
    # function to display delivery information
    def display_delivery_details(self):
        print("Delivery Details")
        print(f"Customer: {self.customer_name}")
        print(f"Address: {self.address}")

customer_name = "Suman"
address = "Hyderabad"
# creating delivery object
delivery = Delivery(customer_name, address)
# displaying delivery details
delivery.display_delivery_details()
```
