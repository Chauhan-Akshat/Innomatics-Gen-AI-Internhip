```python
#Employee Performance Bonus Eligibility

employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

#Finding the highest performance score
highest_score = max(employees.values())

#Finding all employees with the highest score
top_performers = []

for name, score in employees.items():
    if score == highest_score:
        top_performers.append(name)

#Displaying the result
print("Top Performers Eligible for Bonus:", ", ".join(top_performers), f"(Score: {highest_score})")
```

```python
#Search Query Keyword Analysis
import string
query = "Buy mobile phone buy phone online"

#Converting to lowercase
query = query.lower()

#Removing punctuation (if any)
query = query.translate(str.maketrans('', '', string.punctuation))

#Spliting into words
words = query.split()

#Counting frequency
frequency = {}

for word in words:
    frequency[word] = frequency.get(word, 0) + 1

#Displaying words with frequency more than once
result = {}

for word, count in frequency.items():
    if count > 1:
        result[word] = count

print(result)
```

```python
#Sensor Data Validation

sensor_readings = [3, 4, 7, 8, 10, 12, 5]

#Creating list to store valid readings
valid_readings = []

#Checking each reading with its index (hour)
for index, value in enumerate(sensor_readings):
    if value % 2 == 0:   # check if even
        valid_readings.append((index, value))

#Displaying result
print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

```python
#Email Domain Usage Analysis

emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

#Count domains
domain_count = {}

for email in emails:
    domain = email.split("@")[1]
    domain_count[domain] = domain_count.get(domain, 0) + 1

#Calculate percentage
total = len(emails)

for domain, count in domain_count.items():
    percentage = (count / total) * 100
    print(f"{domain}: {int(percentage)}%")
```

```python
#Sales Spike Detection

sales = [1200, 1500, 900, 2200, 1400, 3000]

#Calculate average daily sales
average_sales = sum(sales) / len(sales)

#Find sales more than 30% above average
threshold = average_sales * 1.3

#Display spike days
for i in range(len(sales)):
    if sales[i] > threshold:
        print(f"Day {i+1}: {sales[i]}")
```

```python
#Duplicate User ID Detection

user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

#Count frequency of each user ID
count_dict = {}

for user in user_ids:
    count_dict[user] = count_dict.get(user, 0) + 1

#Display duplicates only
for user, count in count_dict.items():
    if count > 1:
        print(f"{user} → {count} times")
```
